```python
import kagglehub
import pickle
import re
from datasets import load_dataset
import pandas as pd
```

```python
path = kagglehub.dataset_download("perryperryfries/bloomberg-news-articles-sentiment-scores-2006-13")
print("Path to dataset files:", path)
```

```python
bloomberg_articles = pd.read_csv("./bloomberg_news_articles.csv")["Article"]
public_news = load_dataset("vblagoje/cc_news")
```

```python
import numpy as np

public_news_list = list(public_news['train']['text'])
my_set = set()
my_set2 = set()
selected_idx = np.random.choice(len(public_news_list), size=8000, replace=False)  #subsample
for idx in selected_idx:
    my_set.add(public_news_list[idx])
selected_idx = np.random.choice(len(bloomberg_articles), size=8000, replace=False)
for idx in selected_idx:
    my_set2.add(bloomberg_articles[idx])
public_news = pd.DataFrame(my_set)
bloomberg_articles = pd.DataFrame(my_set2)

```

```python
W2V_TOKEN_RE = re.compile(r"[a-z]+(?:'[a-z]+)?")  # keeps words like don't


def tokenize_for_w2v(s: str) -> list[str]:
    s = "" if s is None else str(s)
    s = s.lower()

    # remove urls-ish and email-ish fragments (optional but usually helps)
    s = re.sub(r"\b(?:https?://|www\.)\S+\b", " ", s)
    s = re.sub(r"\b\S+@\S+\b", " ", s)

    # keep alphabetic tokens (drops numbers, punctuation)
    return W2V_TOKEN_RE.findall(s)
```

```python
# Word2Vec training input: list of token lists (one per row)
sentences_bmg = bloomberg_articles.map(tokenize_for_w2v)[0].tolist()

with open("bloomberg_articles.pkl", "wb") as f:
    pickle.dump(sentences_bmg, f, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
# Word2Vec training input: list of token lists (one per row)
sentences_news = public_news.map(tokenize_for_w2v)[0].tolist()

with open("public_news.pkl", "wb") as f:
    pickle.dump(sentences_news, f, protocol=pickle.HIGHEST_PROTOCOL)
```
